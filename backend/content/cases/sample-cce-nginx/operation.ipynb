# 操作步骤（CCE + Nginx）

左侧阅读步骤，右侧 Notebook 执行命令。建议先完成 AK/SK 配置，再执行云资源相关操作。

## Step 1: 准备 kubeconfig 与上下文

1. 获取 CCE 集群访问凭据。
2. 在 Notebook 中写入 `~/.kube/config`。
3. 执行 `kubectl get nodes` 验证连通性。

## Step 2: 发布 Deployment

建议使用 `nginx:stable`，副本数先设为 `2`，随后再体验滚动升级。

```python
kubectl create deployment nginx-demo --image=nginx:stable
kubectl scale deployment nginx-demo --replicas=2
kubectl get pods -o wide
```

## Step 3: 暴露服务并验证

根据集群网络策略选择 `LoadBalancer` 或 `NodePort`。

```python
kubectl expose deployment nginx-demo --port=80 --target-port=80 --type=LoadBalancer
kubectl get svc nginx-demo
kubectl logs deployment/nginx-demo --tail=100
```
